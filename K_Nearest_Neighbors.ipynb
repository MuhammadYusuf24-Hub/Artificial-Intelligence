```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
df = pd.read_csv('Iris.csv')
print(df.head())
```

```python
df = df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species']]
```

```python
le = LabelEncoder()
df['Species'] = le.fit_transform(df['Species'])
```

```python
X = df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y = df['Species']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

print("\nAccuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
plt.scatter(df['PetalLengthCm'], df['PetalWidthCm'], c=df['Species'])
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("Klasifikasi Iris")
plt.colorbar(label="Species")
plt.show()
```

```python
data_baru = [[5.1, 3.5, 1.4, 0.2]]

data_baru_scaled = scaler.transform(data_baru)
pred = model.predict(data_baru_scaled)

print("\nPrediksi jenis bunga:", le.inverse_transform(pred)[0])
```

Analisi Hasil Praktikum

Metode K-Nearest Neighbor (KNN) merupakan algoritma klasifikasi yang bekerja berdasarkan kedekatan jarak antar data. Pada penelitian ini, digunakan dataset Iris yang terdiri dari empat fitur utama, yaitu sepal length, sepal width, petal length, dan petal width, untuk menentukan jenis bunga (species).

Sebelum proses klasifikasi, dilakukan normalisasi data menggunakan StandardScaler. Hal ini penting karena KNN menggunakan perhitungan jarak, sehingga perbedaan skala antar fitur dapat mempengaruhi hasil klasifikasi. Setelah itu, data dibagi menjadi data latih dan data uji, kemudian model dilatih menggunakan nilai K tertentu.

Hasil yang diperoleh menunjukkan bahwa model KNN mampu memberikan akurasi yang tinggi, yang berarti sebagian besar data berhasil diklasifikasikan dengan benar. Berdasarkan visualisasi scatter plot, terlihat bahwa data memiliki pola yang cukup jelas, terutama pada fitur petal length dan petal width. Hal ini membuat algoritma KNN bekerja dengan sangat baik karena jarak antar kelas cukup terpisah.
